```python
import pandas as pd

# Load the dataset
dataset = pd.read_csv("LifeExpectancyData.csv")

# Clean column names by stripping leading and trailing spaces
dataset.columns = dataset.columns.str.strip()

# # Print cleaned column names to verify
# print(dataset.columns.tolist())

# Updated columns to match the cleaned column names
columns_to_replace = [
    'Life expectancy', 'Adult Mortality', 'Alcohol', 'Hepatitis B', 'BMI',
    'Total expenditure', 'Polio', 'Diphtheria', 'GDP', 'Population',
    'thinness  1-19 years', 'thinness 5-9 years', 'Income composition of resources',
    'Schooling'
]

# Replace missing values with 0 for the specified columns
for column in columns_to_replace:
    if column in dataset.columns:
        dataset[column] = dataset[column].fillna(0)
    else:
        print(f"Column '{column}' not found in the dataset")

# Check if all missing values are replaced
print(dataset.isnull().sum())
```

```python
def quanQual(dataset):
        Quan=[]
        Qual=[]
        for columnName in dataset.columns:
            #print(columnName)
            if dataset[columnName].dtypes=='O':
                #print("Qual")
                Qual.append(columnName)
            else:
                #print("Quan")
                Quan.append(columnName)
        return Quan,Qual
```

```python
quan,qual=quanQual(dataset)
```

```python
dataset[quan]
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(dataset[quan])
df=imp.transform(dataset[quan])
```

```python
df
```

```python
df=pd.DataFrame(df,columns=quan)
df
```

```python
category=dataset[qual]
category
```

```python
frames=[df,category]
```

```python
Preprocessed=pd.concat(frames,axis=1)
```
